# Data structures

Data structures are a way of organizing and storing information in a computer or a program. They help in efficiently and manipulating data. 

## Data structures vs an Algorithm 

**An algorithm** is a set of well defined instructions, a step-by-step procedure designed to solve a specific problem or perform a particular task. 

**A data structure** is a way of *organizing* and *storing* data in a computer or a programming language. It defines the relationship between the data elements, the operations that can be performed on the data, and the rules or constraints for accessing and modifying the data. 

*Data structures provide the organization and representation of information
(the data), and algorithms serve as instructions for transforming that data.*


## Array 

Arrays organize data by holding a collection of elements and making them accessible through an index. 

| action   | time    |
| -------- | ------- |
| lookup   | O(1)    |
| push     | O(1)    | 
| insert   | O(n)    |
|  delete  | O(n)    |

*if all you need is to store some data and iterate over it, that is go one by one, step by step this data structure is the best option*

````python
## BINARY SEARCH ALGORITHM

def binary_search (elements:list, item:int):
    low = 0
    high = len(elements)-1
    iterations = 0

    while low <= high :
        mid = int((low+high)/2)
        iterations +=1
        print(iterations)
        guess = elements[mid]
        print("item = ",guess)
        if guess == item: # The guess was too high
            return guess
        if guess > item: # The guess was too low
            high = mid -1
        else: 
            low = mid +1
    return None # The item dows not exist

print(binary_search([1, 3, 5, 7, 9], 1)) # O(logn)
````

|Simple Search  | Binary Search |
| ------------- | ------------- |
| $100$ Items -> $100$ guesses | $100$ Items -> $100$ guesses |
| $4,000 M$ Items-> $4,000 M$ guesses | $4,000 M$ Items-> $32 guesses$ |
| $O(n)$ *Linear Time* | $O(log n)$ *Logarithmic time* |

## Big O notation 
Big O natation tells you how fast an algorithm is. 

**Where are the seconds?** <br>
Big O does not tell you the speed in seconds. Big O notation lets you compare the number of operations. It tells you how fast the algorithm grows. 
>
> ![Big O](img/bigO-notation-example-image.png "BigO-notation")
> Big O notation 

### Common Big O run times 

* $O(log n)$, also known as log time. Example: Binary search.
* $O(n)$, also known as linear time. Example: Simple search.
* $O(n * log n)$. Example: A fast sorting algorithm.
* $O(n^2)$. Example: A slow sorting algorithm. 
* $O(n!)$. Example: A slow sorting algorithm.

| # OF BOXES | $O(logn)$ | $O(n)$ | O(nlogn) | $O(n^2)$ | $O(n!)$ |
|------------|-----------|--------|----------|----------|---------|
| 16  |  0.4s | 1.6s  | 6.4s  | 25.6s  | 66301y  |
| 256  | 0.8s  | 25.6s  | 3.4m  | 1.8h  | $8.6x10^(505) y$  |
| 1024  | 1.0s  | 1.7s | 17m  | 1.2d  | $5.4x10^(2638) y$  |


## RESUME 1 

* Binary search is a lot faster than simple search 
* $O(log n)$ is faster than $O(n)$, but it gets a lot faster once the list of items you're searching through grows. 
* Algorithm speed isn't measured in seconds. 
* Algorithm times are measured in terms of growth of an algorithm 
* Algorithm times are written in **Big O notation**

## What is "Memory", really? 
At the lowest level, memory is just a huge array of bytes. 

Each byte has: 

* an address (a number)
* a value (0-255)

example (conceptual):

```bash
Address     Value
```

**slicing** - slice out substrings, sublists, subtuples using indexes.

````python
strings = ['a', 'b', 'c', 'd']
# storage : 4*4 = 16 bytes of storage

print(strings[2])  #lookup O(1)
strings.append('e') # push O(1)
strings.insert(2, 'e')# insert O(n)
strings.pop() # delete the last element  O(1)
strings.remove('e') # delete O(n)
print(strings)

strings.insert(2, 'o') #O(n)
print(strings)
````

## Array

An array is a linear data structure that stores elements of the same type in contiguous memory locations and allows direct access by index. 

The position of an element is called its *index*. So instead of saying, "20 is at *position* 1", the correct terminology is, "20 is at index 1". 

```rust
Index:  0   1   2   3 
Array: [10, 20, 30, 40]
```


```bash
Action      Value
lookup      O(1)
push        O(1)
insert      O(n)
delete      O(n)
```



| Feature            | Array      | Linked List            |
| ------------------ | ---------- | ---------------------- |
| Memory             | Contiguous | Non-contiguous         |
| Access             | **O(1)**   | O(n)                   |
| Insertion (middle) | O(n)       | O(1) *(if node known)* |
| Cache friendly     | ✅         | ❌                    |


## Linked List

A linked List is a linear data structure where elements (callled nodes) are stored **non-contiguously** in memory

Each item stores the address of the next item in the list. A buch of random memory adresses are linked together.

It's like a treasure hunt. You go to he first address, and says, "The next item can be found at address 847", and so on. Adding an item to a linked list is easy: you stick it anywhere in memory and store the address with the previous item.


1. **Data**: The value being stored
2. **Pointer(link)**: a reference to the next node in the sequence



10 → 20 → 30 → null


```bash
Action      Value
lookup      O(1)
push        O(1)
insert      O(n)
delete      O(n)
```
