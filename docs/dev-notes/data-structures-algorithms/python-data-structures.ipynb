**Stack using List with a Wrapper Class**

We create a Stack class and a full set of Stack methods. <br>
But the underlying data structure is really a Python List. <br>
For pop and peek methods we first check whether the stack is empty, to avoid exceptions.

````python
class Stack():
    def __init__(self):
        self.stack = list()
    def push(self, item):# Add 
        self.stack.append(item)
    def pop(self):# Delete 
        if len(self.stack) > 0:
            return self.stack.pop()
        else: 
            return None
    def peek(self):
        if len(self.stack) > 0:
            return self.stack[-1]
        else: 
            return None
    def is_empty(self):
        if len(self.stack) == 0:
            return True
        else: 
            return False
    def __str__ (self):
        return str(self.stack)
````

## What is a MaxHeap?

* Complete Binary Tree
* Every node $<=$ its parent

**MaxHeap is Fast** <br>
* Insert in $O(log n)$
* Get Max in $O(1)$
* Remove Max in $O(log n)$

> ![MaxHeaps](img\max-heaps.png "max-heaps") <br>
> Max Heaps

Easy to implement using a List <br>
$i = 4$ <br>
$parent(i) = i/2 = 2$ <br>
$left(i) = i*2 = 8$ <br>
$right(i) = i*2 +1 = 9$ <br>

**MaxHeap Operations** <br>
* **Push** (insert)
    * Add value to end of array
    * Float it Up to its proper position
* **Peek**(get max)
    * Return the value at heap
* **Pop** (remove max)
    * Move max to end of array
    * Delete it 
    * Buble Down the item at index 1 to its proper position
    * Return Max



**Python MaxHeap** <br>
A maxHeap always bubbles the highest value to the top, so it can be removed instantly. <br>
Public functions: push, peed, pop <br>
Private functions: swap, __floatUP, bubbleDown, __str. 


````python
class MaxHeap: 
    def __init__(self, items=[]):
        super().__init__()
        self.heap = [0]
        for item in items: 
            self.heap.append(item)
            self.__floatUp(len(self.heap)-1)
    
    def push (self, data): 
        self.heap.append(data)
        self.__floatUp(len(self.heap)-1)
    
    def peek(self): 
        if self.heap[1]:
            return self.heap[1]
        else: 
            return False
    
    def pop(self): 
        if len(self.heap) > 2: 
            self.__swap(1, len(self.heap)-1)
    
    def __swap(self, i, j): 
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]
    def __floatUp(self, index):
        parent = index//2
        if index <= 1: 
            return 
        elif self.heap[index] > self.heap[parent]:
            self.__swap(index, parent)
            self.__floatUp(parent)
````

## Linked List 

Every Node has 2 parts. <br>
**data** and a pointer to he **next** Node

**Attributes:**
* **root** - pointer to the beginning of the List 
* **size** - number of nodes in List 

**Operations:** 
* find(data)
* add(data)
    * To add a new node we add that before the **root node**, so the next node will be converted in the root node.
* remove(data)
    * To remove a node, the node behind change its pointer to the node after the node to be deleted.  
* print_list()

````python
class Node: 
    def __init__(self, d, n=None, p=None):
        self.data = d
        self.next_node = n 
        self.prev_node = p
    
    def __str__(self): 
        return (f"({str(self.data)})")
````

````python
class LinkedList: 

    def __init__(self, r = None): 
        self.root = r 
        self.size = 0
    
    def add(self, d): 
        new_node = Node(d, self.root) # Creating a new node
        self.root = new_node # Assigning a new root
        self.size += 1 
    
    def find(self, d): 
        this_node = self.root
        while this_node is not None: 
            if this_node.data == d: 
                return d 
            else: 
                this_node = this_node.next_node
    
    def remove(self, d): 
        this_node = self.root
        prev_node = None

        while this_node is not None:
            if this_node.data == d:
                if prev_node is not None: # data in non-root
                    prev_node.next_node = this_node.next_node
                else: # data is in root node
                    self.root = this_node.next_node
                self.size -=1
                return True # data removed
            else: 
                prev_node = this_node
                this_node = this_node.next_node
        return False # data not found
    
    def print_list(self): 
        this_node = self.root 
        while this_node is not None: 
            print(this_node, end='->')
            this_node = this_node.next_node
        print('None')
````

## Linked List Comprehension 


````python
from typing import List, Iterable, Union
class Node: 
    def __init__(self, val):
        self.val = val 
        self.next = None

a = Node('A')
b = Node('B')
c = Node('C')
d = Node('D')

a.next = b 
b.next = c 
c.next = d 
d.next = None 

# A -> B -> C -> D -> None 
#              head

# print a linked_list 
def print_linked_list(root:Node)->str: 
    if not root: return "-1"
    current = root
    ls = []
    while current: 
        ls.append(current.val)
        current = current.next 
    return "->".join(map(str, ls))

def print_linked_list_r(root:Node, ls: List = [])->str: 
    if not root: 
        if ls == []: 
            return "-1"
        return "->".join(map(str,ls))
    ls.append(root.val)
    return print_linked_list_r(root.next, ls)

print(print_linked_list(a)) # output A->B->C->D
print(print_linked_list_r(a)) # output A->B->C->D 
````

````python
# sum Linked List

def sum_linked_list(root:Node) -> int | None: 
    if not root: return None 
    val = 0
    current = root
    while current: 
        val += current.val
        current = current.next
    return val

def sum_linked_list_r(root:Node) -> int:
    if not root:
        return 0  
    return root.val + sum_linked_list_r(root.next)

a = Node(1)
b = Node(2)
c = Node(3)
d = Node(4)

a.next = b 
b.next = c 
c.next = d 
d.next = None

print(sum_linked_list(a)) # Output: 10 
print(sum_linked_list_r(a)) # Output: 10  
````

````python
# Linked List Find 

def find_linked_list_val(root:Node, target) -> bool:
    if not root: return False
    current = root 
    while current: 
        if current.val == target: 
            return True 
        current = current.next
    return False

def find_linked_list_val_r(root: Node, target): 
    if not root: return False 
    elif root.val == target: 
        return True 
    return find_linked_list_val(root.next, target)

print(find_linked_list_val(a, 5)) # Output: False
print(find_linked_list_val_r(a, 3)) # Output: True
````

````python
# get node value 

def get_node_val(root:Node, index:int) -> int | None: 
    if not root: return None 
    current = root 
    count = 0 
    while current: 
        if count == index: 
            return current.val 
        count += 1
        current = current.next
    return None 

def get_node_val_r(root:Node, index:int) -> int | None:
    if not root: return None 
    elif index  == 0: return root.val
    return get_node_val_r(root.next, index-1)

print(get_node_val(a, 15)) 
print(get_node_val(a, 2)) # Output: 3
````

````python
# Reverse List 

def reverse_linked_list(head: Node | None) -> Node | None:
    prev: Node | None = None
    cur: Node | None = head

    while cur:
        nxt = cur.next
        cur.next = prev
        prev = cur
        cur = nxt

    return prev

def reverse_linked_list_r(head: Node, prev: Node | None = None): 
    if not head: return prev # The End condition 
    nx = head.next 
    head.next = prev 
    return reverse_linked_list_r(nx, head)

a = Node("A")
b = Node("B")
c = Node("C")
d = Node("D")

a.next = b 
b.next = c 
c.next = d 
d.next = None


#   a  ->  b  ->  c
#  curr 
print(print_linked_list(a))# Output: A -> B -> C -> D
print(print_linked_list(reverse_linked_list_r(a))) # Output: D -> C -> B -> A
````

````python
# Zipper List
# Explaining: 
# A -> B -> C 
# Q -> R -> S 
# result: A -> Q -> B -> R -> C -> S 

def zipper_list(head1:Node, head2:Node) -> Node | None: 
    tail = head1
    current1 = head1.next
    current2 = head2 
    count = 0
    while current1 and current2: 
        if count % 2 == 0: 
            tail.next = current2
            current2 = current2.next
        else: 
            tail.next = current1
            current1 = current1.next 
        
        tail = tail.next
        count += 1
    
    if current1: tail.next = current1 
    elif current2: tail.next = current2
    return head1


def zipper_list_r(head1:Node, head2:Node) -> Node:
    if not head1 and not head2: 
        return None 
    if not head1: return head2
    if not head2: return head1
    
    nx1 = head1.next 
    nx2 = head2.next
    
    head1.next = head2
    head2.next =  zipper_list_r(nx1, nx2)
    return head1


a = Node("A")
b = Node("B")
c = Node("C")
d = Node("D")

a.next = b 
b.next = c 
c.next = d 
d.next = None

e = Node("E")
f = Node("F")
g = Node("G")
h = Node("H")

e.next = f 
f.next = g 
g.next = h 
h.next = None

print(print_linked_list(zipper_list_r(a, e)))# Output A->E->B->F->C->G->D->H
````

## Circular Linked List 
**Advantage** over regular (singly) linked lists <br>
* Ideal for modeling continuous looping objects, such as Monopoly board or a race track. 


````python
class CircularLinkedList: 

    def __init__(self, r = None): 
        self.root = r 
        self.size = 0
    
    def add(self, d): 
        if self.size == 0: 
            self.root = Node(d)
            self.root.next_node = self.root
        else: 
            new_node = Node(d, self.root.next_node)
            self.root.next_node = new_node
        self.size += 1
    
    def find (self, d): 
        this_node = self.root
        while True: 
            if this_node.data == d: 
                return d 
            elif this_node.next_node == self.root: 
                return False
            this_node = this_node.next_node
        
    def remove (self, d): 
        this_node = self.root
        prev_node = None

        while True: 
            if this_node.data == d: # found
                if prev_node is not None: 
                    prev_node.next_node = this_node.next_node
                else: 
                    while this_node.next_node != self.root: 
                        this_node = this_node.next_node
                    this_node.next_node = self.root.next_node 
                    self.root = self.root.next_node
                self.size-=1
                return True # data removed
            elif this_node.next_node == self.root: 
                return False  # data not found 
            prev_node = this_node 
            this_node = this_node.next_node
        
    def print_list(self): 
        if self.root is None: 
            return 
        this_node = self.root 
        print(this_node, end='->')
        while this_node.next_node != self.root: 
            this_node = this_node.next_node
            print(this_node, end='->')
        print()
````

## Doubly Linked List

Every Node has 3 parts: <br>
**data** and pointers to **previous** and **next** Nodes

**Advantages** over regular (singly) linked list 
* Can iterate the list in either direction
* Can delete a node without iterating through the list (if given a pointer to the node) 

````python
class DoublyLinkedList: 
    def __init__ (self): 
        self.root = None
        self.last = None 
        self.size = 0
    
    def add(self, d): 
        if self.size == 0: 
            self.root = Node(d)
            self.last = self.root # None
        else: 
            new_node = Node(d, self.root)
            self.root.prev_node = new_node
            self.root = new_node
        self.size +=1
    
    def find(self, d): 
        this_node = self.root
        while this_node is not None: 
            if this_node.data == d: 
                return d 
            elif this_node.next_node == None: 
                return False
            else: 
                this_node = this_node.next_node
    
    def remove (self, d): 
        this_node = self.root 
        while this_node is not None: 
            if this_node.data == d: 
                if this_node.prev_node is not None: 
                    if this_node.next_node is not None: 
                        this_node.prev_node.next_node = this_node.next_node
                        this_node.next_node.prev_node = this_node.prev_node
                    else: 
                        this_node.prev_node.next_node = None
                        self.last = this_node.prev_node
                else: # delete root node 
                    self.root = this_node.next_node
                    this_node.next_node.prev_node = self.root
                self.size-=1
                return True # data removed
            else: 
                this_node = this_node.next_node
        return False # data not found 
    
    def print_list(self): 
        if self.root is None: 
            return 
        this_node = self.root
        print(this_node, end="->")
        while this_node.next_node is not None: 
            this_node = this_node.next_node
            print(this_node, end='->')
        print('None')
````

## Binary Search Tree 

**BST Characteristics** 
* At most 2 children per node 
* Exactly 1 root
* 1 Path between root and any node


**BST Operations**
* Insert 
    * Start at root 
    * Always insert as a leaf
* Find 
    * Start at root 
    * Return the data if found, or False if not found
* Delete (3 possible cases)
    * leaf node 
        * Just delete the leaf node
    * 1 child 
        * Promote the child to the target node's position
    * 2 children
        * Find the next higher node, change the place and delete the leaf
* Get_size
    * Returns number of nodes. 
    * Works recursively
* Traversals 
    * Visit root before visiting the root's subtrees. 
    * Visit root between visiting the root’s subtrees. 
Gives values in sorted order.

**Advantages of Binary Trees?** <br>
* Because trees use recursion for most operations. they are fairly easy to implement. 
* Speed: Insert, Delete, Find in $O(h) = O(log n)$
* In a balance BST with 10,000,000 nodes It'd takes 30 comparisons!

### Binary Search Tree Class
**Constructor** sets three attributes: data, left subtree and right subtree. <br> 
**Insert** inserts a new subtree into the proper location. <br> 
**Find** finds a value. If value not found, returns False. <br>
**Preorder** prints a preorder traversal of the tree. <br>
**Inorder** prints an inorder traversal of the tree. 

````python
class Tree: 
    def __init__(self, data, left=None, right=None):
        self.data = data 
        self.left = left 
        self.right = right 
    def insert(self, data):
        if self.data == data: 
            return False # duplicate value 
        elif self.data > data: 
            if self.left is not None: 
                return self.left.insert(data)
            else: 
                self.left = Tree(data)
                return True 
        else: 
            if self.right is not None: 
                return self.right.insert(data)
            else: 
                self.right = Tree(data)
                return True
    def find(self, data): 
        if self.data == data: 
            return data
        elif self.data>data: 
            if self.left is None: 
                return False
            else:
                return self.left.find(data)
        elif self.data < data: 
            if self.right is None: 
                return False
            else: 
                return self.right.find(data)
    def get_size(self):
        if self.left is not None and self.right is not None: 
            return 1 + self.left.get_size() + self.right.get_size()
        elif self.left: 
            return 1 + self.left.get_size()
        elif self.right: 
            return 1 + self.right.get_size()
        else: return 1
    def preorder(self): 
        if self is not None: 
            print(self.data, end=" ")
            if self.left is not None: 
                self.left.preorder()
            if self.right: 
                self.right.preorder()
    def inorder(self): 
        if self is not None: 
            if self.left is not None: 
                self.left.inorder()
            print(self.data, end=' ')
            if self.right is not None: 
                self.right.inorder()
````

### DFS (Depth First Search)

````python
def depth_first_value(root: Node) -> List:
    if root == None: return []
    stack = [root]
    result = []
    while len(stack) > 0: 
        current = stack.pop()
        result.append(current.val)
        if current.right: 
            stack.append(current.right)
        if current.left: 
            stack.append(current.left)
    return (result)

def depth_first_value2(root: Node) -> List:  
    if root == None: return []
    leftValues = depth_first_value2(root.left)
    rightValues = depth_first_value2(root.right)

    return [root.val, leftValues, rightValues]

def treeIncludes(root:Node, target) -> bool: 
    if root == None: return False 
    if root.val == target: return True
    return treeIncludes(root.left, target) or treeIncludes(root.right, target)

print(depth_first_value(a))
print(print(treeIncludes(a, 'f')))
# Output: ['a', 'b', 'd', 'e', 'c', 'f']
# True 
````

### BFS (Breadth First Search)

````python
def breadth_first_search(root: Node) -> List: 
    if root == None: return []

    my_queue = deque()
    my_queue.append(root)
    val = []

    while my_queue:
        current = my_queue.popleft() #O(1)
        val.append(current.val)
        if current.left: 
            my_queue.append(current.left)
        if current.right: 
            my_queue.append(current.right)
    return val
    
print(breadth_first_search(a))
# Output: ['a', 'b', 'c', 'd', 'e', 'f']
````

````python
# Three min_val 

def three_min_val(root: Node)->int: 
    if not root: return float('inf')
    my_queue = deque()
    my_queue.append(root)
    min_val = None 
    while my_queue: 
        current = my_queue.popleft()
        if min_val == None: 
            min_val = current.val
        elif min_val > current.val: 
            min_val = current.val
        
        if current.left: my_queue.append(current.left)
        if current.right: my_queue.append(current.right)
    return min_val

def three_min_val2(root: Node)->int: 
    if not root: return float('inf') # or math.inf 
    left_min = three_min_val2(root.left)
    right_min = three_min_val2(root.right)
    return min(root.val, left_min, right_min)

a2 = Node(5)
b2 = Node(11)
c2 = Node(3)
d2 = Node(4)
e2 = Node(15)
f2 = Node(12)

a2.left = b2
a2.right = c2
b2.left = d2
b2.right = e2
c2.right = f2

print(three_min_val(a2)) #O(n)
print(three_min_val2(a2)) #O(n)
````

````python
# Island Count 

def island_count(grid): 
    visited = set()
    count = 0 
    for r in range(len(grid)): 
        for c in range(len(grid[0])): 
            if explore_ocean(grid, r, c, visited): 
                count += 1
    return count

def explore_ocean(grid, row, column, visited): # recursive DFS solution
    row_in_bounds = (0 <= row and row < len(grid)) # True or False 
    col_in_bounds = (0 <= column and column < len(grid[0])) # True or False
    
    if not row_in_bounds or not col_in_bounds: 
        return False 
    
    if grid[row][column] == 'W': 
        return False

    position = (row, column)  
    if position in visited: return False 
    visited.add(position)

    explore_ocean(grid, row-1, column, visited) # left 
    explore_ocean(grid, row+1, column, visited) # right
    explore_ocean(grid, row, column-1, visited) # down 
    explore_ocean(grid, row, column+1, visited) # up

    return True 

grid = [
    ['W', 'L', 'W', 'W', 'W'], 
    ['W', 'L', 'W', 'W', 'W'], 
    ['W', 'W', 'W', 'L', 'W'],
    ['W', 'W', 'L', 'L', 'W'],
    ['L', 'W', 'W', 'L', 'L'],
    ['L', 'L', 'W', 'W', 'W']
]

print(island_count(grid))
````

````python
# Palindrome 
# kayak 

def is_palindrome(word): 
    left, right = 0, len(word)-1
    while right >= left:
        if word[left] != word[right]: 
            return False 
        left += 1
        right -= 1
    return True 


def is_palindrome_r(word): 
    # Define the base-case / stopping condition
    if len(word) == 0 or len(word) == 1: 
        return True 
    # Do some work to shrink the problem space
    if word[0] == word[-1]:
        return is_palindrome_r(word[1:-1])
    # Additional base-case to handle non-palindromes
    return False

palindrome = 'kayak'
print(is_palindrome_r('racecar'))
````

**Divide and Conquer**

1. Divide problem into several smaller subproblems Normally, the subproblems are similar to the original.
2. Conquer the subproblems by solving them recursively. Base case: solve small enough problems by brute force. 
3. Combine the solutions to get a solution to the subproblems and finally a solution to the original problem. 
4. Divide and Conquer algorithms are normally recursive.

````python
# Binary Search Recursive

def binary_search(ls, left, right, target):
    if right <= left: 
        return -1
    
    mid = (left + right)//2

    if ls[mid] == target: 
        return mid 
    
    if target < ls[mid]:
        return binary_search(ls, left, mid-1, target)
    return binary_search(ls, left, left, mid+1)
````

````python
# Fibonacci (Non-Optimized)

def fib(n): # (Non-Optimized)
    if n == 0 or n == 1: 
        return n
    return fib(n-1) + fib(n-2)
# * Each recursive call is stacked on the call stack 
# * Nothing is added until a base case is reached 
# * Then the stack is unstacked and the operations are executed upon return 

# Spanish: 
# * Cada llamada recursiva se aplica en la pila de llamadas 
# * Nada se suma hasta que se alcanza un caso base 
# * Luego el stack se desapila y las operaciones se ejecutan al regresar

def fib2(n):
    left, right = 1, 1
    middle = 0 
    for _ in range(n-2):
        middle = left + right 
        left = right 
        right = middle
    return right


print(fib2(10))
````

````python
## Fibonacci Problem 
## Write a function 'fib(n)' that takes in a number as an argument. 
## The function should return the n-th number of the Fibonacci sequence. 

# Time: O(2^n)
# Space: O(n)
def fib(n): # Not optimized fibonacci function
    if n <= 2: return 1
    return fib(n-1) + fib(n-2)

# Time O(n)
# Space O(n)
def fib_optimized(n, memo = dict()): 
    if n <= 2: return 1
    if n in memo: return memo[n]
    memo[n] = fib_optimized(n-1, memo) + fib_optimized(n-2, memo)
    return memo[n]  

def fib_optimized2(n):
    if n == 1 or n == 2: 
        return 1
    ans = 2 # Third response 
    last = 1
    last_last = 1
    for i in range(3, n + 1): 
        ans = last + last_last
        last_last = last 
        last = ans
    return ans 

print(fib_optimized(7)) # Output: 13
print(fib_optimized(7))# Output: 13
print(fib_optimized(8))# Output: 21
print(fib_optimized(50))# Output: 12586269025
print(fib_optimized2(7))# Output: 13
````

````python
# Say that you are a traveler on a 2D grid. You begin in the top-left corner
# and your goal is to travel to the bottom-right corner. 
# You may only move down or right
# 
#  In how many ways can you travel to the goal on a grid with dimensions m*n? 

def grid_traveler(m, n): 
    if m == 1 and n == 1: return 1
    if m == 0 or n == 0: return 0 # empty 
    return grid_traveler(m-1, n) + grid_traveler(m, n -1)


def grid_traveler_optimized(m, n, memo= None):
    if memo == None: 
        memo = {}
    
    key = (m, n)
    if key in memo: 
        return memo[key]
    if m == 1 and n == 1: return 1 
    if m == 0 or n == 0: return 0
    memo[key] = grid_traveler_optimized(m-1, n, memo) + grid_traveler_optimized(m, n-1, memo)
    return memo[key]
print(grid_traveler(1, 1)) # Output: 1 (rows, columns)

print(grid_traveler(2, 3)) # Output: 3
print(grid_traveler(15, 4))# Output: 6

print(grid_traveler_optimized(15, 4)) # Output: 3
````

````python
# canSum Problem: 

def can_sum(target:int, numbers:list)->list: 
    # m = target sum 
    # n = array length
    # T: O(n^m)
    # S: O(m)
    if target == 0: 
        return True 
    
    if target < 0: 
        return False
    
    for num in numbers: 
        remainder = target - num
        if can_sum(remainder, numbers) == True: 
            return True
    return False

def can_sum_optimized(target:int, numbers: list, memo = None): 
    # T: O(m * n) 
    # S: O(m)
    if memo == None: 
        memo = {}
    
    if target in memo: 
        return memo[target]

    if target == 0: 
        return True
    
    if target < 0: 
        return False 
    
    for i in numbers: 
        remainder = target - i 
        if can_sum_optimized(remainder, numbers, memo): 
            memo[target] = True
            return True 
    
    memo[target] = False
    return False 


print(can_sum(7,[2, 3])) # True
print(can_sum(7,[5, 3, 4, 7])) # True
print(can_sum(7,[ 2, 4])) # False
print(can_sum(8, [2, 3, 5])) # True 
# print(can_sum(300, [7, 14])) # False Not efficient in can_sum

print('Optimized UwU')
print(can_sum_optimized(7,[2, 3])) # True
print(can_sum_optimized(7,[5, 3, 4, 7])) # True
print(can_sum_optimized(7,[ 2, 4])) # False
print(can_sum_optimized(8, [2, 3, 5])) # True 
print(can_sum_optimized(300, [7, 14])) # False 
````

````python
# How sum problem 

def how_sum(target, numbers): 
    # m -> target 
    # n = len(numbers)
    # T: O(n^m * m)
    # S: O(m)
    if target == 0: 
        return []
    
    if target < 0: 
        return None 
    
    for num in numbers: 
        remainder = target - num
        remainder_result = how_sum(remainder, numbers)
        if remainder_result != None: 
            return remainder_result + [num]
    
    return None

def how_sum_optimized(target, numbers, memo = None): 
    # T: O(n*m^2)
    # S: O(m^2)
    if memo == None: 
        memo = dict()
    
    if target in memo: 
        return memo[target]

    if target == 0: 
        return []

    if target  < 0: 
        return None 
    
    for num in numbers: 
        remainder = target - num 
        remainder_number = how_sum_optimized(remainder, numbers, memo)
        if remainder_number != None:
            memo[target] = remainder_number + [num]
            return memo[target]
    
    memo[target] = None
    return None

# proving how_sum
print(how_sum(7, [2, 3]))
print(how_sum(7, [5, 3, 4, 7]))
print(how_sum(7, [2, 4]))
print(how_sum(8, [2, 3, 5]))
#print(how_sum(300, [7, 14])) # Not efficient 

# proving how_sum_optimized
print(how_sum_optimized(7, [2, 3]))
print(how_sum_optimized(7, [5, 3, 4, 7]))
print(how_sum_optimized(7, [2, 4]))
print(how_sum_optimized(8, [2, 3, 5]))
print(how_sum_optimized(300, [7, 14]))
````

````python
# best sum

# Brute force solution
def best_sum(target: int, numbers: list): 
    # T: O(n^m *m)
    # S: O(m^2)
    if target == 0: 
        return []
    
    if target < 0: 
        return None
    
    shortest_combination = None

    for num in numbers:
        remainder = target - num 
        remainder_combination = best_sum(remainder, numbers)
        if remainder_combination != None: 
            combination = remainder_combination + [num]
            # if the combination is shortest than the current shortest
            if shortest_combination == None or len(combination) < len(shortest_combination): 
                shortest_combination = combination

    return shortest_combination

# Memoization Solution
def best_sum_optimized(target: int, numbers: list, memo: None|dict = None)-> list|None:
    # T: O(n*m^2)
    # S: O(m^2)
    if memo == None: 
        memo = dict()
    
    if target in memo: 
        return memo[target]

    if target == 0: 
        return []
    
    if target < 0: 
        return None
    
    if target in memo: 
        return memo[target]
    
    shortest_combination = None 

    for num in numbers: 
        remainder = target - num
        remainder_combination = best_sum_optimized(remainder, numbers, memo)
        if remainder_combination != None: 
            combination = remainder_combination + [num]
            if shortest_combination == None or len(combination) < len(shortest_combination): 
                shortest_combination = combination

    memo[target] = shortest_combination 
    return shortest_combination


print(best_sum(7, [5, 3, 4, 7]))# 7
print(best_sum(8, [2, 3, 5])) # 5, 3
print(best_sum(8, [1, 4, 5])) # 4, 4
# print(best_sum(100, [1, 2, 5, 25])) # Not optimized

print('--------')
print(best_sum_optimized(7, [5, 3, 4, 7]))
print(best_sum_optimized(8, [2, 3, 5]))
print(best_sum_optimized(8, [1, 4, 5]))
print(best_sum_optimized(100, [1, 2, 5, 25]))
````

````python
# Can Construct? 
# Not optimized version
def can_construct(target: str, word_bank: list):
    # m = len(target)
    # n = len(word_bank)
    # Time: O(n^m * m)
    # Space: O(m)
    if target == '': # The True condition 
        return True 
    
    for word in word_bank:
        if target.startswith(word): 
            suffix = target[len(word):] # Creating the new target:
            if can_construct(suffix, word_bank) == True: 
                return True
    return False 
# binary 
#   


# Optimized version
def can_construct_optimized(target: str, word_bank: list, memo:None|dict = None):
    # Time: O(m^2*n)
    # Space: O(m)
    if not memo: 
        memo = {}
    
    if target == '': # The True condition 
        return True 
    
    for word in word_bank:
        if target.startswith(word): 
            suffix = target[len(word):] # Creating the new target:
            if can_construct_optimized(suffix, word_bank) == True: 
                memo[target] = True 
                return memo[target]
    
    memo[target] = False
    return memo[target] 

print(can_construct('abcdef', ['ab', 'abc', 'cd', 'def', 'abcd'])) # True 
print(can_construct('skateboard', ['bo', 'rd', 'ate', 't', 'ska', 'sk', 'boar'])) # False
print(can_construct('enterapotentpot', ['a', 'p', 'ent', 'enter', 'ot', 'o', 't'])) # True 

print('-----------')

print(can_construct_optimized('abcdef', ['ab', 'abc', 'cd', 'def', 'abcd'])) # True 
print(can_construct_optimized('skateboard', ['bo', 'rd', 'ate', 't', 'ska', 'sk', 'boar'])) # False
print(can_construct_optimized('enterapotentpot', ['a', 'p', 'ent', 'enter', 'ot', 'o', 't'])) # True 


````

````python
# count_ construct 

def count_construct(target: str, word_bank: list) -> int:
    # Time: O(n^m * m)
    # Space: O(m^2)
    if target == '':
        return 1
    
    total_count = 0

    for word in word_bank:
        if target.startswith(word): 
            suffix = target[len(word):]
            num_ways_for_rest = count_construct(suffix, word_bank)
            total_count += num_ways_for_rest
    
    return total_count

def count_construct_optimized(target:str, word_bank:list, memo = None) -> int: 
    # Time: O(n* m^2)
    # Space: O(m^2)
    if memo == None: 
        memo = {}
    
    if target == '': 
        return 1
    
    if target in memo: 
        return memo[target]
    
    total_count = 0 
    for word in word_bank: 
        if target.startswith(word): 
            suffix = target[len(word):]
            num_ways_for_rest = count_construct_optimized(suffix, word_bank, memo)
            total_count += num_ways_for_rest
    memo[target] = total_count
    return memo[target] 


print(count_construct('purple', ['purp', 'p', 'ur', 'le', 'purpl'])) # 2
print(count_construct('abcdef', ['ab', 'abc', 'cd', 'def', 'abcd'])) # 1 
print(count_construct('skateboard', ['bo', 'rd', 'ate', 't', 'ska', 'sk', 'boar'])) # 0
print(count_construct('enterapotentpot', ['a', 'p', 'ent', 'enter', 'ot', 'o', 't'])) # 4 
print("----------- + ----------- + -----------")
print(count_construct_optimized('purple', ['purp', 'p', 'ur', 'le', 'purpl'])) # 2
print(count_construct_optimized('abcdef', ['ab', 'abc', 'cd', 'def', 'abcd'])) # 1 
print(count_construct_optimized('skateboard', ['bo', 'rd', 'ate', 't', 'ska', 'sk', 'boar'])) # 0
print(count_construct_optimized('enterapotentpot', ['a', 'p', 'ent', 'enter', 'ot', 'o', 't'])) # 4 
````

````python
def all_construct(target: str, word_bank: list[str]) -> list[list[str]]:
    if target == "":
        return [[]]

    result = []

    for word in word_bank:
        if target.startswith(word):
            suffix = target[len(word):]
            suffix_ways = all_construct(suffix, word_bank)
            target_ways = [[word] + suffix_way for suffix_way in suffix_ways]
            result.extend(target_ways)

    return result

def all_construct_optimized(target: str, word_bank: list[str], memo=None):
    if memo is None:
        memo = {}

    if target in memo:
        return memo[target]

    if target == "":
        return [[]]

    result = []

    for word in word_bank:
        if target.startswith(word):
            suffix = target[len(word):]
            suffix_ways = all_construct_optimized(suffix, word_bank, memo)
            target_ways = [[word] + suffix_way for suffix_way in suffix_ways]
            result.extend(target_ways)

    memo[target] = result
    return result


print(all_construct('purple', ['purp', 'p', 'ur', 'le', 'purpl']))


print(all_construct_optimized('purple', ['purp', 'p', 'ur', 'le', 'purpl']))
````
